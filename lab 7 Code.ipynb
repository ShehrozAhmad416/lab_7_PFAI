```python
#task no.1
import seaborn as sns
import pandas as pd
tips = sns.load_dataset('tips')
print(tips.head())
print("Rows and Columns:", tips.shape)
print(tips.describe())
```

```python
#task no.2
import matplotlib.pyplot as plt
sns.histplot(data=tips, x='tip', kde=True, color='skyblue')
plt.title('Distribution of Tip Amounts')
plt.xlabel('Tip ($)')
plt.ylabel('Count')
plt.show() 
```

```python
#task no.3
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='sex',color='blue')
plt.title('Total Bill vs Tip by Time and Gender')
plt.xlabel('Total Bill ($)',fontsize=15)
plt.ylabel('Tip ($)',fontsize=15)
plt.grid('true',alpha=0.9)
plt.show()
```

```python
#task no.4
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
plt.title('Total Bill by Day and Smoker Status')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```

```python
#task no.5
corr = tips[['total_bill','tip','size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Numerical Variables')
plt.show()
```
